```python
from googletrans import Translator
```

```python
import pandas as pd
```

```python
df = pd.read_csv('marathi_tweets.csv')
```

```python
df.values.tolist()[0]
```

```python
translator = Translator()
translatedlist = [translator.translate(tweet[0], dest="en").text for tweet in df.values.tolist()]
```

```python
print(translatedlist)
```

```python
tweets = pd.DataFrame(translatedlist)
```

```python
tweets.to_csv('translated_tweets.csv',index=False)
```
